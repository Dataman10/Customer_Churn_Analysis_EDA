```python
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Group and Get Counts
data = df.groupby(['Partner', 'Churn']).size().unstack(fill_value=0)

# Step 2: Convert to Row-Wise Percentage
percent = data.div(data.sum(axis=1), axis=0) * 100

# Step 3: Plot Stacked Bar
plt.figure(figsize=(4,4))
ax = percent.plot(
    kind='bar',
    stacked=True,
    figsize=(4,4),
    colormap="Set2"
)

plt.title("Churn Percentage by Partner")
plt.xlabel("Partner")
plt.ylabel("Percentage (%)")
plt.legend(title="Churn")

# Step 4: Add % Labels Inside the Bars
for i, row in enumerate(percent.values):
    cumulative = 0
    for val in row:
        if val > 0:
            plt.text(
                i,
                cumulative + val / 2,
                f"{val:.1f}%",
                ha='center',
                va='center',
                fontsize=9,
                color='black'
            )
        cumulative += val

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Create subplots (3 rows x 3 columns)
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()   # 2D array → 1D for easy looping

# Loop through columns and plot
for i, col in enumerate(cols):
    sns.countplot(
        x=col,
        data=df,
        ax=axes[i],
        hue = df['Churn']
    )
    axes[i].set_title(f"Count of Customers by {col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Count")
    
    # Add bar labels
    for container in axes[i].containers:
        axes[i].bar_label(container)

plt.tight_layout()
plt.show()
```

Most customers have PhoneService enabled, while a significant portion does not use MultipleLines, indicating limited add-on adoption.
For internet-related services, Fiber optic and DSL users dominate, but many customers fall under “No internet service”, which directly impacts features like OnlineSecurity, OnlineBackup, DeviceProtection, and TechSupport.
Streaming services (TV & Movies) show a balanced adoption, suggesting they are optional rather than essential services.
Overall, customers with fewer bundled services appear more common, which may correlate with higher churn risk and presents opportunities for upselling bundled plans
